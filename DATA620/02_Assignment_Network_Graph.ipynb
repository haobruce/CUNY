```python
o=nx.Graph()

o.add_edge('Conrad','Mary')
o.add_edge('Conrad','Cindy')
o.add_edge('Conrad','Alice')
o.add_edge('Alice','Brad')
o.add_edge('Alice','Angie')
o.add_edge('Alice','Jim')
o.add_edge('Cindy','Samuel')
o.add_edge('Cindy','Dave')
o.add_edge('Cindy','Frida')
nx.draw(o)
```

```python
G=nx.Graph()

labels={}
labels[0]='Conrad'
labels[1]='Mary'
labels[2]='Cindy'
labels[3]='Alice'
labels[4]='Brad'
labels[5]='Angie'
labels[6]='Jim'
labels[7]='Samuel'
labels[8]='Dave'
labels[9]='Frida'

G.add_nodes_from(range(0,10))  # add 10 nodes 
# color coding nodes based on gender 
color_map=[]
for node in G: 
    if node in [0,4,6,7,8]: 
        color_map.append('blue')
    else: 
        color_map.append('red')

G.add_edge(0,1)
G.add_edge(0,2)
G.add_edge(0,3)
G.add_edge(3,4)
G.add_edge(3,5)
G.add_edge(3,6)
G.add_edge(2,7)
G.add_edge(2,8)
G.add_edge(2,9)

# create position after defining nodes and edges 
pos=nx.spring_layout(G)

# draw nodes, edges and labels 
nx.draw_networkx_nodes(G,pos,node_color=color_map,alpha=0.5)
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
# add colored coded edges for same-gender edges 
nx.draw_networkx_edges(G,pos,
                       edgelist=[(3,5),(2,9)],
                       width=8,alpha=0.5,edge_color='r')
nx.draw_networkx_labels(G,pos,labels,font_size=16)
```

For Project 1, I built a simple websraper to look at social network data on Facebook. I applied the snowball method starting with one central node and went three layers deep. I sampled just the first page of friends for each node to keep things manageable. I then trimmed the graph for nodes with less than 10 degrees and then gathered attribute data for the remaining nodes.
```python
%matplotlib inline
import os
import networkx as net
import pandas as pd
```

```python
# load network graph data and show graph 
g = net.read_pajek(os.getcwd() + '/_data/facebook_network.net')

def show_network(g3):
    pos = net.spring_layout(g3)
    net.draw_networkx_nodes(g3, pos, alpha=0.5)
    net.draw_networkx_edges(g3, pos, alpha=0.5)
    net.draw_networkx_labels(g3, pos, alpha=0.7)

show_network(g)
```

```python
# compute centrality measures: degree, closeness, betweenness and eigenvector closeness 
def sorted_map(map):
    ms = sorted(map.items(), key=lambda kv: kv[1], reverse=True)
    return ms
```

```python
# degree
d = net.degree(g)
ds = sorted(g.degree, key=lambda x: x[1], reverse=True)
ds_df = pd.DataFrame.from_records(ds[:10], columns=['user_name', 'degree'])
ds_df['degree']  # hiding username for privacy purposes 
```

```python
# closeness 
c = net.closeness_centrality(g)
cs = sorted_map(c)
cs_df = pd.DataFrame.from_records(cs[:10], columns=['user_name', 'closeness'])
cs_df['closeness']  # hiding username for privacy purposes 
```

```python
# betweenness 
b = net.betweenness_centrality(g)
bs = sorted_map(b)
bs_df = pd.DataFrame.from_records(bs[:10], columns=['user_name', 'betweenness'])
bs_df['betweenness']  # hiding username for privacy purposes 
```

```python
e = net.eigenvector_centrality_numpy(g)
ec = sorted_map(e)
ec_df = pd.DataFrame.from_records(ec[:10], columns=['user_name', 'eigenvector'])
ec_df['eigenvector']  # hiding username for privacy purposes 
```

```python
# print table of users with highest centrality measures 
names1=[x[0] for x in ds[:10]]
names2=[x[0] for x in cs[:10]]
names3=[x[0] for x in bs[:10]]
names4=[x[0] for x in ec[:10]]
names=list(set(names1) | set(names2) | set(names3) | set(names4))
table=[[name, d[name], c[name], b[name], e[name]] for name in names]
df = pd.DataFrame(table, columns=['user_name','degree','closeness','betweenness','eigenvector'])

# sum ranks to order by centrality - assumes equal weighting 
df['rank'] = df.rank(axis=0, ascending=False).sum(axis=1)
df.drop('user_name', axis=1, inplace=True)  
df.sort_values(by=['rank'])
```

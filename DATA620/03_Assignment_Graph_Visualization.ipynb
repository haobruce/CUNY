```python
%matplotlib inline
import networkx as nx
import networkx.generators.small
```

```python
# replicating the example in the book so that I understand how it works 
g = networkx.generators.small.krackhardt_kite_graph()
print('number of edges: %d' % g.number_of_edges())
print('number of nodes: %d' % g.number_of_nodes())
```

```python
pos=nx.spring_layout(g)
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_nodes(g, pos)
nx.draw_networkx_labels(g, pos)
```

```python
print('diameter: %d' % nx.diameter(g))
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_nodes(g, pos)
nx.draw_networkx_labels(g, pos)
# add colored edge to illustrate one possible diameter 
nx.draw_networkx_edges(g,pos,
                       edgelist=[(9,8),(8,7),(7,5),(5,2)],
                       width=8,alpha=0.5,edge_color='b')
```

```python
# the eccentricity of a node is the maximum distance to any other node in the graph 
print(nx.eccentricity(g))
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_nodes(g, pos)
nx.draw_networkx_labels(g, pos)
# add colored edge to illustrate eccentricity for node 0
nx.draw_networkx_edges(g,pos,
                       edgelist=[(0,5),(5,7),(7,8),(8,9)],
                       width=8,alpha=0.5,edge_color='b')
```

```python
# the center is the node that achieves the minimum eccentricty 
print(nx.center(g))
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_nodes(g, pos)
nx.draw_networkx_labels(g, pos)
# add color to highlight center node 
nx.draw_networkx_nodes(g, pos, 
                       nodelist=[7],
                       node_color='b')
```
